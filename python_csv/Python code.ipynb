```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import lxml
```

```python
url = "https://en.wikipedia.org/wiki/Schengen_Area" 
heads = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'}
response = requests.get(url, headers = heads)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, "lxml")
```

```python
table = soup.find("table", class_= "sortable")
```

```python
headers = [th.text.strip() for th in table.find_all("th")]
rows = []
for row in table.find_all("tr")[1:]:
    cells = [td.text.strip() for td in row.find_all("td")]
    rows.append(cells)
```

```python
df = pd.DataFrame({"Table": [table]})
```

```python
df.to_csv("Summary_table.csv", index = False)
```
